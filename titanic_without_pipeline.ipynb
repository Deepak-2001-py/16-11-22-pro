```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('/storage/emulated/0/Download/train.txt', delimiter =',')
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
df.head()
```

```python
df.shape
```

```python
#step-1 - - >train _test_split
X_train, X_test, y_train, y_test=train_test_split(df.drop(['Survived'], axis=1),df['Survived'] , test_size =0.2,random_state =42)
```

```python
X_train.head(2)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
# Applying Imputation
si_age= SimpleImputer()
si_embarked=SimpleImputer(strategy='most_frequent')

X_train_age=si_age.fit_transform(X_train[['Age']])
X_train_embarked=si_embarked.fit_transform(X_train[['Embarked']])

X_test_age=si_age.fit_transform(X_test[['Age']])
X_test_embarked=si_embarked.fit_transform(X_test[['Embarked']])

```

```python
X_train_embarked.shape,X_test_embarked.shape
```

```python
X_train[['Sex']]
```

```python
X_train_embarked
```

```python
# One Hot Encoding
ohe_sex= OneHotEncoder(sparse=False, handle_unknown='ignore' )
ohe_embarked=OneHotEncoder(sparse=False, handle_unknown='ignore')

X_train_sex=ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked=ohe_embarked.fit_transform(X_train_embarked)  

X_test_sex=ohe_sex.fit_transform(X_test[['Sex']])
X_test_embarked=ohe_embarked.fit_transform(X_test_embarked)
```

```python
X_train_embarked
```

```python
X_train_embarked.shape, X_test_embarked.shape
```

```python
X_train.head(2)
```

```python
X_train_rem=X_train.drop(columns=['Sex','Age','Embarked'], axis=1)
X_test_rem=X_test.drop(columns=['Sex','Age','Embarked'], axis=1)                             
```

```python
X_train_transform=np.concatenate((X_train_rem,X_train_sex,X_train_age,X_train_embarked),axis=1)
X_test_transform=np.concatenate((X_test_rem,X_test_sex,X_test_age,X_test_embarked),axis=1)
```

```python
X_test_transform.shape
```

```python
pd.DataFrame(X_train_transform) 
```

```python
pd.DataFrame(X_test_transform) 
```

```python
clf=DecisionTreeClassifier()
clf.fit(X_train_transform,y_train)
```

```python
y_pred=clf.predict(X_test_transform)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred) 
```
